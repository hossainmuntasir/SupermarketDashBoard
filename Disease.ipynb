```python
import pandas as pd
import matplotlib.pyplot as plt
disease = pd.read_csv("Disease-2.csv")
print(disease)
```

```python
plt.plot(disease["Day"],disease["Cases"], color = "g")
plt.title("Cases per day", fontsize =20, color = "red" )
plt.xlabel("Day",fontsize= 15 , color = "b")
plt.ylabel("Cases per day", fontsize = 15, color = "b")
```
