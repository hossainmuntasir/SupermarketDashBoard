```python
import pandas as pd
import matplotlib.pyplot as plt
salary = pd.read_excel("Acc Salaries-1.xlsx","Python")
print(salary)
```

```python
freq = salary["Salaries"].value_counts(bins = 6, sort = False)
print("Groupped frequency data")
print(freq)

#plt.hist(salary["Salaries"],bins = 6 , color = "g")
plt.hist(salary["Salaries"],bins =[40, 66, 92, 118, 144, 170, 196, 222], color="blue")
plt.title("Q5 salaries", fontsize = 20, color ="red")
plt.xlabel("salaries",fontsize = 25, color = "orange")
plt.ylabel("frequency", fontsize = 25, color = "orange")
```
